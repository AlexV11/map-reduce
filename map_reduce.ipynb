```python
import time

# función que cuenta las palabras en un archivo y las agrega a un diccionario
def count_words(filename, word_count):
    with open(filename, 'r', encoding="utf8") as file:
        for line in file:
            words = line.split()
            for word in words:
                if word not in word_count:
                    word_count[word] = 0
                word_count[word] += 1

# función que suma las frecuencias de las palabras en todo el conjunto de datos
def reduce_word_counts(word_counts):
    word_count = {}
    for wc in word_counts:
        for word, count in wc.items():
            if word not in word_count:
                word_count[word] = 0
            word_count[word] += count
    return word_count

# función principal que divide los datos y realiza el conteo de palabras
def main():
    start_time = time.time()
    filenames = ['archivo1.txt', 'archivo2.txt', 'archivo3.txt']
    word_counts = []
    for filename in filenames:
        wc = {}
        count_words(filename, wc)
        word_counts.append(wc)
    word_count = reduce_word_counts(word_counts)
    print(word_count)
    end_time = time.time()
    print("Tiempo de ejecución: ", end_time - start_time)

if __name__ == '__main__':
    main()
```

Procesameinto paralelo

```python
import threading
import time

# función que se ejecuta en cada hilo para contar las palabras en un archivo
def count_words(filename, word_count):
    with open(filename, 'r', encoding="utf8") as file:
        for line in file:
            words = line.split()
            for word in words:
                if word not in word_count:
                    word_count[word] = 0
                word_count[word] += 1

# función de reducción que suma las frecuencias de las palabras en todo el conjunto de datos
def reduce_word_counts(word_counts):
    word_count = {}
    for wc in word_counts:
        for word, count in wc.items():
            if word not in word_count:
                word_count[word] = 0
            word_count[word] += count
    return word_count

# función principal que divide los datos en hilos y realiza el conteo de palabras
def main():
    start_time = time.time()
    filenames = ['archivo1.txt', 'archivo2.txt', 'archivo3.txt']
    word_counts = []
    threads = []
    for filename in filenames:
        wc = {}
        word_counts.append(wc)
        t = threading.Thread(target=count_words, args=(filename, wc))
        t.start()
        threads.append(t)
    for t in threads:
        t.join()
    word_count = reduce_word_counts(word_counts)
    print(word_count)
    end_time = time.time()
    print("Tiempo de ejecución: ", end_time - start_time)

if __name__ == '__main__':
    main()
```
